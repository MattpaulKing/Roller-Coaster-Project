```python
###Creating a function that ranks a given rollercoaster's ranking over time

def coaster_plot(name, park, df):
    coaster = df[(df.Name == name) & (df.Park == park)]
    x_values = coaster["Year of Rank"].values
    y_values = coaster["Rank"].values

    ax = plt.subplot()
    ax.plot(x_values, y_values, label = coaster.Name)
    ax.set_yticks(range(len(y_values)))
    plt.ylabel("Ranking")
    ax.invert_yaxis()
    plt.xlabel("Year of Ranking")
    plt.title("Ranking of " + name)
    plt.show()
```

```python
coaster_plot("El Toro", "Six Flags Great Adventure", wood)
```

```python
###Writing a function that plots the ranking of two roller coasters over time

def coaster_plot2(name1, name2, df, park1, park2):
    coaster1 = df[(df.Name == name1) & (df.Park == park1)]
    coaster2 = df[(df.Name == name2) & (df.Park == park2)]
    x_values1 = coaster1["Year of Rank"].values
    x_values2 = coaster2["Year of Rank"].values
    y_values1 = coaster1["Rank"].values
    y_values2 = coaster2["Rank"].values
    
    ax = plt.subplot()
    plt.plot(x_values1, y_values1, label = name1)
    plt.plot(x_values2, y_values2, label = name2)
    ax.set_yticks(range(len(y_values1)))
    plt.ylabel("Ranking")
    ax.invert_yaxis()
    plt.xlabel("Year of Ranking")
    plt.title("Ranking of " + name1 + " & " + name2)
    plt.legend()
    plt.show()
    
    
```

```python
coaster_plot2("El Toro", "Boulder Dash", wood, "Six Flags Great Adventure", "Lake Compounce")
```

```python
### Writing a function that will plot the top ranking of the top n rated roller coasters over time

def nrankedcoaster(n, coaster_df):
    ###Setting the df to only include the top n ranked coasters
    df = coaster_df[coaster_df['Rank'] <= n]
    ax = plt.subplot()
    ### Iterating through each roller coaster by name (Using set(df.Name) to make the series iterable)
    for roller_coaster in set(df.Name):
        ###Creating a new df so that I can plot each coaster separately
        roller_df = df[df.Name == roller_coaster]
        ax.plot(roller_df["Year of Rank"], roller_df["Rank"], label=roller_coaster, marker=0)
    ax.invert_yaxis()
    plt.xlabel("Year of Ranking")
    plt.ylabel("Rank")
    plt.title("Roller Coasters in top " + str(n) + " Rankings")
    ### Making the legend appear on the outside of the plot so that the graph is easily Readable
    plt.legend(loc = "center left", bbox_to_anchor=(1.0, 0.5))
    ax.set_yticks(range(1, n+1))
    plt.figure(figsize=(20, 16))
    plt.show()
        
            
```

```python
nrankedcoaster(10, wood)
```

```python
### Opening the informational roller coaster CSV and exploring the data
df_coasters = pd.read_csv("roller_coasters.csv")
print(df_coasters.head())
print(df_coasters.info())
```

```python
###Plotting the numerical columns as a histogram
def coaster_hist(df, colname):
    ax = plt.subplot()
    plt.hist(df[colname], label=colname)
    plt.title(colname)
    plt.xlabel(colname)
    plt.ylabel("Count of Coasters")

    
```

```python
coaster_hist(df_coasters, 'speed')
```

```python
### Creating bar graphs for the number of inversions for a specific roller coaster at a park
def coasterbar(df, park_name):
    park_coasters = df[df['park'] == park_name]
    
    ## As a rule of thumb getting the x and y values out of the way cleans up the code
    x_values = park_coasters.name
    y_values = park_coasters.num_inversions.dropna()
    
    ## Making the plot + Cleaning up the titles/labels
    ax = plt.subplot()
    ax.bar(range(len(x_values)), y_values)
    ax.set_xticks(range(len(x_values)))
    ax.set_xticklabels(x_values, rotation=90)
    plt.xlabel("Coaster Names",)
    plt.ylabel("Number of Inversions")
    plt.title("Number of Inversions per Coaster at " + park_name)
    plt.show()
    
                          
```

```python
coasterbar(df_coasters, 'Terra Mítica')
```

```python
### Creating a plot that outputs the PCT of operating roller coasters to the number of non-operating

def coasterpie(df):
    operating = df[df['status'] == 'status.operating']
    non_operating = df[df['status'] == 'status.closed.temporarily']
    status_counts = [len(operating), len(non_operating)]
    
    ax=plt.subplot()
    ax.pie(status_counts, labels=('Operating', "Non Operating"), autopct='%d%%')
    plt.show()
```

```python
coasterpie(df_coasters)
```

```python
### Creating a scatter plot of any two numerical columns

def coasterscatter(df, column1, column2):
    points1 = df[column1].values
    points2 = df[column2].values
    
    ax = plt.subplot()
    plt.scatter(points1, points2)
    plt.xlabel(column1)
    plt.ylabel(column2)
    plt.title("{} vs {}".format(column1, column2))
    plt.show()
```

```python
coasterscatter(df_coasters, 'speed', 'length')
```
